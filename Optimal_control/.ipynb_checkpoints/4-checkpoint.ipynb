```python
from model import *
from src.problem.prb import Analytic

device = torch.device('cuda') if torch.cuda.is_available() else torch.device('cpu')
an = Analytic(device) 

act_func=lambda x: torch.tanh(x)

V_NN = V_Net(dim=2, ns=128, act_func=act_func, hh=0.5,
                            device=device, psi_func=an.psi_func, TT=1).to(device)
    
V_NN.load_state_dict(torch.load('V_NN.pth'))

```

```python
num_points=500

x = create_meshgrid_2d(num_points)
    
with torch.no_grad():
    V_NN.eval()
    # t = 0
    t = 0*torch.ones(num_points*num_points,1, device=device)
    v_pred = V_NN(t,x)
    v_exact = an.V_exact(x,t=0)
    err = relative_error(v_exact, v_pred)
    print("Relative_Error_t0 = ", err)

    # t = 0.25
    t = 0.25*torch.ones(num_points*num_points,1, device=device)
    v_pred = V_NN(t,x)
    v_exact = an.V_exact(x,t=0.25)
    err = relative_error(v_exact, v_pred)
    print("Relative_Error_t0.25 = ", err)

    # t = 0.5
    t = 0.5*torch.ones(num_points*num_points,1, device=device)
    v_pred = V_NN(t,x)
    v_exact = an.V_exact(x,t=0.5)
    err = relative_error(v_exact, v_pred)
    print("Relative_Error_t0.5 = ", err)
    
    # t = 0.75
    t = 0.75*torch.ones(num_points*num_points,1, device=device)
    v_pred = V_NN(t,x)
    v_exact = an.V_exact(x,t=0.75)
    err = relative_error(v_exact, v_pred)
    print("Relative_Error_t0.75 = ", err)

    # t = 1
    t = torch.ones(num_points*num_points,1, device=device)
    v_pred = V_NN(t,x)
    v_exact = an.V_exact(x,t=1.0)
    err = relative_error(v_exact, v_pred)
    print("Relative_Error_tf = ", err)
```

```python
def V_exact(x, t):
    return -2*(x/(1+ math.exp(2*(t-1)))).view(-1,1)

num_points=1000
x = torch.linspace(-1, 1, num_points, device=device).unsqueeze(1)


#with torch.no_grad():
#V_NN.eval()
# t = 0
t = 0*torch.ones(num_points,1, device=device)
v_pred = -1/2*V_NN.get_grad(t,x)
#np.savez('U_IGT_t=0.npz', u=v_pred.cpu().detach().numpy())
v_exact = V_exact(x,t=0)
err = relative_error(v_exact, v_pred)
print("Relative_Error_t0 = ", err)

# t = 0.25
t = 0.25*torch.ones(num_points,1, device=device)
v_pred = -1/2*V_NN.get_grad(t,x)

v_exact = V_exact(x,t=0.25)
err = relative_error(v_exact, v_pred)
print("Relative_Error_t0.25 = ", err)

# t = 0.5
t = 0.5*torch.ones(num_points,1, device=device)
v_pred = -1/2*V_NN.get_grad(t,x)
#np.savez('U_IGT_t=0,5.npz', u=v_pred.cpu().detach().numpy())
v_exact = V_exact(x,t=0.5)
err = relative_error(v_exact, v_pred)
print("Relative_Error_t0.5 = ", err)

# t = 0.75
t = 0.75*torch.ones(num_points,1, device=device)
v_pred = -1/2*V_NN.get_grad(t,x)
v_exact = V_exact(x,t=0.75)
err = relative_error(v_exact, v_pred)
print("Relative_Error_t0.75 = ", err)

# t = 1
t = torch.ones(num_points,1, device=device)
v_pred = -1/2*V_NN.get_grad(t,x)
#np.savez('U_IGT_t=1.npz', u=v_pred.cpu().detach().numpy())
v_exact = V_exact(x,t=1.0)
err = relative_error(v_exact, v_pred)
print("Relative_Error_tf = ", err)
```

```python
import matplotlib.pyplot as plt
import numpy as np
# Plotting t =0
V_NNA0 = np.load('V_NNA_t=0.npz')
V_NNA0 = V_NNA0['v']

V_ASMR0 = np.load('V_ASMR_t=0.npz')
V_ASMR0 = V_ASMR0['v']



num_points = 100
t = 0*torch.ones(num_points,1, device=device)
x = torch.linspace(-1, 1, num_points, device=device).unsqueeze(1)


err_IGT_0 = 6.4788e-5
err_ASMR_0 = 0.003657005
err_NNA_0 = 0.0181028

v_exact = an.V_exact(x,t=0.)

with torch.no_grad():
    V_NN.eval()
    IGT = V_NN(t,x)

plt.figure(figsize=(6, 4))

plt.plot(x.cpu(), v_exact.cpu(), 'g', label='Exact Solution', linestyle='-')
plt.plot(x.cpu(), IGT.cpu(), 'b-.', label='IGT Approximation')
plt.plot(x.cpu(), V_ASMR0.T, 'y--', label='ASMR Approximation')
plt.plot(x.cpu(), V_NNA0, 'r:', label='NNA Approximation')



# Annotate relative errors
plt.text(0.5, 0.5, f'Relative_Error_IGT: {err_IGT_0:.2e}\nRelative_Error_ASMR: {err_ASMR_0:.2e}\nRelative_Error_NNA: {err_NNA_0:.2e}',
         fontsize=10, ha='center', va='center', transform=plt.gca().transAxes)

#plt.title('Comparison of Approximated Solutions with Exact Solution')
plt.xlabel('x')
plt.ylabel('v')
plt.title('t=0')
plt.legend()
plt.grid(True)
plt.grid(color='gray', linestyle='--', linewidth=0.35)
plt.savefig('v0.png', dpi=300)
plt.show()
```
