```python
from IPython.display import clear_output
import numpy as np
from scipy.integrate import solve_ivp, solve_bvp

def M0(x):
  m = torch.exp(-((x[:,1]) ** 2) / 0.001)
  mean = torch.sum(m)
  test = (x[:, 0] >= -0.02) & (x[:,0] <= 0.02)
  return ((1/(0.04) * test.int() * m)/sum((1/(0.04) * test.int() * m))).view(-1,1)

def M00(x):
        m = np.exp(-((x[:,1]) ** 2) / 0.001)
        mean = np.sum(m)
        condition = np.logical_and(x[:, 0] >= -0.02, x[:, 0] <= 0.02)
        condition = condition.astype(np.float32)  # Convert boolean to float
        return (1/(0.04) * condition * (m))/sum(1/(0.04) * condition * (m))#/ mean


def gen_x0( num_samples):

    samples = []

    while len(samples) < num_samples:
        # Generate random samples from a uniform distribution
        x_candidate = 0.6*np.random.uniform(size=(num_samples, 2))-0.3

        # Evaluate the density for the candidate samples
        prob_candidate = M00(x_candidate)

        # Generate random values to accept/reject the samples
        acceptance_prob = np.random.uniform(size=num_samples)

        # Accept samples where the acceptance probability is greater than the density value
        accepted_samples = x_candidate[acceptance_prob < prob_candidate]

        # Append the accepted samples to the result
        samples.extend(accepted_samples.tolist())

    # Trim the excess samples
    x0 = np.array(samples[:num_samples])
    
    return x0

    
def dynamics( t, X, U_fun):

    U = U_fun([[t]], X.reshape((1,-1))).flatten()
    dxdt = U
    dydt = X[:1]
    
    return np.concatenate((dxdt,dydt))

def eval_u(t, x):
    
        u = - V_NN.get_grad(t, x)[:,0].detach().cpu().numpy()
        
        return u
    


dim = 2

X_OUT = np.empty((0, dim))
t_OUT = np.empty((0, 1))

Ns_sol = 0
num_samples = 512
x0_int = gen_x0(num_samples)
# data = np.load('data.npz')
# x0_int = data['x']

while Ns_sol < num_samples:
    
        #clear_output(wait=True)
        print('Solving BVP #', Ns_sol+1, '...', end='\r')

        X0 = x0_int[Ns_sol,:]
        
        # Integrates the closed-loop system (NN controller)

        SOL = solve_ivp(dynamics, [0., 1], X0,
                        method= 'RK23', t_eval=np.linspace(0,1,21),
                        args=(eval_u,),
                        rtol=1e-08)

        

        Ns_sol += 1
        
        t_OUT = np.vstack((t_OUT, SOL.t.reshape(1,-1).T))
        X_OUT = np.vstack((X_OUT, SOL.y.T))
        
np.savez('sol_x.npz', X_OUT=X_OUT)        
```

```python
import torch
import torch.nn.functional as F
import matplotlib.pyplot as plt

def total_variation_distance(P, Q, grid_resolution=10, range_x=(-2, 4), range_y=(-2, 6)):
    """
    Compute the Total Variation Distance (TVD) between two distributions of points in R^2.

    Parameters:
    - P (torch.Tensor): Tensor of shape (N, 2) representing distribution P in R^2.
    - Q (torch.Tensor): Tensor of shape (M, 2) representing distribution Q in R^2.
    - grid_resolution (int): Number of bins along each dimension for discretization.
    - range_x (tuple): Range of x-axis for histogram bins.
    - range_y (tuple): Range of y-axis for histogram bins.

    Returns:
    - float: Total Variation Distance between distributions P and Q.
    """
    # Compute histograms for P and Q
    hist_P, x_bins, y_bins = compute_histogram(P, grid_resolution, range_x, range_y)
    hist_Q, _, _ = compute_histogram(Q, grid_resolution, range_x, range_y)

    # Flatten histograms and compute probabilities
    hist_P_flat = hist_P.flatten() / torch.sum(hist_P)
    hist_Q_flat = hist_Q.flatten() / torch.sum(hist_Q)

    # Total Variation Distance (TVD)
    tvd = 0.5 * torch.sum(torch.abs(hist_P_flat - hist_Q_flat))
    
    # Plot histograms
    plot_histogram(hist_P, x_bins, y_bins, title="Histogram of Distribution P")
    plot_histogram(hist_Q, x_bins, y_bins, title="Histogram of Distribution Q")
    
    return tvd.item()  # Convert to Python float

def compute_histogram(points, grid_resolution, range_x, range_y):
    """
    Compute a 2D histogram for a set of points.

    Parameters:
    - points (torch.Tensor): Tensor of shape (N, 2) representing points in R^2.
    - grid_resolution (int): Number of bins along each dimension for discretization.
    - range_x (tuple): Range of x-axis for histogram bins.
    - range_y (tuple): Range of y-axis for histogram bins.

    Returns:
    - torch.Tensor: 2D histogram tensor.
    - torch.Tensor: Bins for x-axis.
    - torch.Tensor: Bins for y-axis.
    """
    # Extract points coordinates
    x_coords = points[:, 0]
    y_coords = points[:, 1]

    # Define bins for histogram
    x_bins = torch.linspace(range_x[0], range_x[1], grid_resolution + 1)
    y_bins = torch.linspace(range_y[0], range_y[1], grid_resolution + 1)

    # Compute indices for each point
    x_indices = torch.bucketize(x_coords, x_bins)
    y_indices = torch.bucketize(y_coords, y_bins)

    # Compute 2D histogram
    hist = torch.zeros(grid_resolution, grid_resolution)
    for i in range(len(points)):
        hist[x_indices[i]-1, y_indices[i]-1] += 1

    return hist, x_bins, y_bins

def plot_histogram(hist, x_bins, y_bins, title="Histogram"):
    """
    Plot a 2D histogram.

    Parameters:
    - hist (torch.Tensor): 2D histogram tensor.
    - x_bins (torch.Tensor): Bins for x-axis.
    - y_bins (torch.Tensor): Bins for y-axis.
    - title (str): Title for the plot.
    """
    plt.figure(figsize=(8, 6))
    plt.imshow(hist.T, extent=(x_bins[0], x_bins[-1], y_bins[0], y_bins[-1]), origin='lower', cmap='Blues')
    plt.colorbar(label='Frequency')
    plt.xlabel('X')
    plt.ylabel('Y')
    plt.title(title)
    plt.grid(True)
    plt.show()

# Example usage:
if __name__ == "__main__":
    P = torch.tensor([[1.0, 2.0], [3.0, 4.0], [2.0, 5.0], [1.0, 1.0], [0.0, 0.0]])
    Q = torch.tensor([[0.0, 1.0], [3.0, 3.0], [2.0, 4.0], [-1.0, 0.0]])

    tvd_distance = total_variation_distance(P, Q, grid_resolution=1000, range_x=(-2, 4), range_y=(-2, 6))
    print("Total Variation Distance (TVD):", tvd_distance)
```

```python
import torch
import numpy as np
from scipy.stats import ks_2samp

def kolmogorov_smirnov_test(P, Q):
    """
    Perform Kolmogorov-Smirnov test between two distributions P and Q.

    Parameters:
    - P (torch.Tensor or np.ndarray): Sample data from distribution P.
    - Q (torch.Tensor or np.ndarray): Sample data from distribution Q.

    Returns:
    - float: Kolmogorov-Smirnov statistic D.
    - float: Two-tailed p-value.
    """
    if isinstance(P, torch.Tensor):
        P = P.cpu().detach().numpy()
    if isinstance(Q, torch.Tensor):
        Q = Q.cpu().detach().numpy()

    D, p_value = ks_2samp(P.flatten(), Q.flatten())
    return D, p_value

# Example usage:
if __name__ == "__main__":
    # Generate example data (random normal distributions)
    P = torch.randn(1000,2)  # Distribution P
    Q = torch.randn(1000,2)  # Distribution Q

    # Perform Kolmogorov-Smirnov test
    D, p_value = kolmogorov_smirnov_test(P, Q)

    print(f"Kolmogorov-Smirnov statistic D: {D:.4f}")
    print(f"Two-tailed p-value: {p_value:.4f}")
```

```python
import numpy as np

def kl_divergence(P, Q, bins=100):
    """
    Compute KL divergence between two empirical distributions P and Q.

    Parameters:
    - P (np.ndarray): Sample data from distribution P.
    - Q (np.ndarray): Sample data from distribution Q.
    - bins (int): Number of bins for histogram estimation.

    Returns:
    - float: KL divergence.
    """
    # Compute histograms
    hist_P, _ = np.histogram(P, bins=bins, density=True)
    hist_Q, _ = np.histogram(Q, bins=bins, density=True)

    # Handle zero probabilities to avoid NaN in KL divergence
    hist_P = np.clip(hist_P, 1e-12, None)
    hist_Q = np.clip(hist_Q, 1e-12, None)

    # Compute KL divergence
    kl_div = np.sum(hist_P * np.log(hist_P / hist_Q))

    return kl_div

# Example usage:
if __name__ == "__main__":
    # Generate example data (random normal distributions)
    P = np.random.randn(1000)  # Distribution P
    Q = np.random.randn(1000)  # Distribution Q

    # Compute KL divergence
    kl = kl_divergence(P, Q)

    print(f"KL Divergence: {kl:.4f}")
```

```python
import numpy as np

def kl_divergence(P, Q, bins=100):
    """
    Compute KL divergence between two empirical distributions P and Q,
    normalized by the number of points.

    Parameters:
    - P (np.ndarray): Sample data from distribution P.
    - Q (np.ndarray): Sample data from distribution Q.
    - bins (int): Number of bins for histogram estimation.

    Returns:
    - float: Normalized KL divergence.
    """
    # Compute histograms
    hist_P, _ = np.histogram(P, bins=bins, density=True)
    hist_Q, _ = np.histogram(Q, bins=bins, density=True)

    # Handle zero probabilities to avoid NaN in KL divergence
    hist_P = np.clip(hist_P, 1e-12, None)
    hist_Q = np.clip(hist_Q, 1e-12, None)

    # Compute KL divergence normalized by number of points
    N = len(P)
    kl_div = np.sum(hist_P * np.log(hist_P / hist_Q)) / N

    return kl_div

# Example usage:
if __name__ == "__main__":
    # Generate example data (random normal distributions)
    P = np.random.randn(1000)  # Distribution P
    Q = np.random.randn(1000)  # Distribution Q

    # Compute normalized KL divergence
    kl = kl_divergence(P, Q)

    print(f"Normalized KL Divergence: {kl:.4f}")
```

```python
import numpy as np
from scipy.optimize import linprog
from scipy.stats import norm

# Set random seed for reproducibility
np.random.seed(42)

# Number of samples
n = 1000
m = 1000

# Generate samples from two normal distributions
mu1, sigma1 = 0, 0.5  # Mean and standard deviation for distribution P
mu2, sigma2 = 0, 0.5  # Mean and standard deviation for distribution Q

P_samples = np.random.normal(mu1, sigma1, n)
Q_samples = np.random.normal(mu2, sigma2, m)

# Compute the pairwise distance matrix (using L1 norm for simplicity)
D = np.abs(P_samples[:, None] - Q_samples[None, :])

# Objective function (flattened distance matrix)
c = D.flatten()

# Constraints
A_eq = []
b_eq = []

# Each point in P should map to exactly 1/n of the mass
for i in range(n):
    A = np.zeros((n, m))
    A[i, :] = 1
    A_eq.append(A.flatten())
    b_eq.append(1/n)

# Each point in Q should map to exactly 1/m of the mass
for j in range(m):
    A = np.zeros((n, m))
    A[:, j] = 1
    A_eq.append(A.flatten())
    b_eq.append(1/m)

A_eq = np.array(A_eq)
b_eq = np.array(b_eq)

# Bounds for the transportation plan variables
x_bounds = [(0, None)] * (n * m)

# Solve the linear program
result = linprog(c, A_eq=A_eq, b_eq=b_eq, bounds=x_bounds, method='highs')

# Optimal transport cost (Wasserstein distance)
W1 = result.fun
print(f'1-Wasserstein distance: {W1}')
```

```python
import torch
import torch.nn as nn
import torch.optim as optim
from torchvision import datasets, transforms
from torch.utils.data import DataLoader

# Define a simple neural network
class SimpleNN(nn.Module):
    def __init__(self):
        super(SimpleNN, self).__init__()
        self.fc1 = nn.Linear(28*28, 512)
        self.fc2 = nn.Linear(512, 512)
        self.fc3 = nn.Linear(512, 10)
        self.relu = nn.ReLU()

    def forward(self, x):
        x = x.view(-1, 28*28)
        x = self.relu(self.fc1(x))
        x = self.relu(self.fc2(x))
        x = self.fc3(x)
        return x

# Load MNIST dataset
transform = transforms.Compose([
    transforms.ToTensor(),
    transforms.Normalize((0.1307,), (0.3081,))
])

train_dataset = datasets.MNIST(root='./data', train=True, transform=transform, download=True)
test_dataset = datasets.MNIST(root='./data', train=False, transform=transform, download=True)

train_loader = DataLoader(train_dataset, batch_size=128, shuffle=True)
test_loader = DataLoader(test_dataset, batch_size=1000, shuffle=False)

# Function to train the network
def train_network(model, train_loader, epochs=5):
    criterion = nn.CrossEntropyLoss()
    optimizer = optim.Adam(model.parameters())
    for epoch in range(epochs):
        model.train()
        for data, target in train_loader:
            optimizer.zero_grad()
            output = model(data)
            loss = criterion(output, target)
            loss.backward()
            optimizer.step()

# Update function
def update_M(NN1, NN2, n):
    for param1, param2 in zip(NN1.parameters(), NN2.parameters()):
        param1.data = (1 / (n + 1)) * param1.data + (n / (n + 1)) * param2.data
    return NN1

# Create two instances of SimpleNN
NN1 = SimpleNN()
NN2 = SimpleNN()

# Train NN1
train_network(NN1, train_loader, epochs=5)

# Train NN2 (similar setup as NN1)
train_network(NN2, train_loader, epochs=5)

# Update NN1 with parameters averaged with NN2
n = 5
NN1 = update_M(NN1, NN2, n)

# Function to evaluate the network
def evaluate_model(model, test_loader):
    model.eval()
    correct = 0
    total = 0
    with torch.no_grad():
        for data, target in test_loader:
            output = model(data)
            _, predicted = torch.max(output, 1)
            total += target.size(0)
            correct += (predicted == target).sum().item()
    return correct / total

# Evaluate the combined network (NN1 after update)
combined_accuracy = evaluate_model(NN1, test_loader)
print(f'Combined network accuracy: {combined_accuracy:.4f}')
```

```python
import torch, numpy as np, copy
from geomloss import SamplesLoss          # pip install geomloss
from model import *

def gen_x00(num_samples):

    samples = np.random.normal(loc=0.1, scale=np.sqrt(0.105), size=(num_samples, 1))

    return samples
    
x0_ini = gen_x00(1000)
x0_ini = torch.tensor(x0_ini, dtype=torch.float32, device=device)
mu  = x0_ini.mean(0);   std = x0_ini.std(0)
G_cfg = dict(dim=1, ns=128, act_func=act_func, hh=0.5,
             device=device, mu=mu, std=std, TT=1)


def wasserstein_fp(G_NN_list,              # BR passés (longueur k)
                   G_NN_new,               # BR courant   (G_k)
                   x0, m0,                    # (N,d)  échantillon initial
                   t_grid,                 # (T,1)  instants
                   p          = 1,
                   blur       = 0.5,
                   device     = "cpu"):
    """
    W_p(  m̄_k(t) , BR_k(t) )  pour t ∈ t_grid,   avec
        m̄_k = (1/(k+1)) [ δ_{x0} + Σ_{i=0}^{k-1} δ_{G_i} ].
    Retourne  np.ndarray shape (T,)  des distances.
    """

    T       = t_grid.shape[0]
    N, d    = x0.shape
    k       = len(G_NN_list)
    loss_fn = SamplesLoss("sinkhorn", p=p, blur=blur, backend="tensorized")

    # ---------- pré-calcul des trajectoires BR_k -------------------
    with torch.no_grad():
        t_big = t_grid.repeat_interleave(N, dim=0)          # (T*N,1)
        x_rep = x0.repeat(T, 1)                             # (T*N,d)
        br_k  = G_NN_new(t_big, x_rep).view(T, N, d)        # (T,N,d)

    # ---------- pré-calcul des trajectoires passées ----------------
    past_traj = []                                          # liste (k,T,N,d)
    for G in G_NN_list:
        with torch.no_grad():
            past = G(t_big, x_rep).view(T, N, d)
        past_traj.append(past)
    if past_traj:
        past_traj = torch.stack(past_traj, dim=0)           # (k,T,N,d)

    # ---------- distances -----------------------------------------
    dists = torch.empty(T, device="cpu")
    for j in range(T):
        # population : x0 + BR passés (chacun N pts)
        parts = [m0]                                        # (N,d)
        if k:
            parts.append(past_traj[:, j, :, :].reshape(k*N, d))
        X = torch.cat(parts, dim=0)                         # ((k+1)*N,d)
        Y = br_k[j]                                         # (N,d)
        dists[j] = loss_fn(X, Y).cpu()

    return dists.numpy()        # shape (T,)
```

```python
import copy, time
import numpy as np
import torch
from model import *
from geomloss import SamplesLoss                 # pip install geomloss

# ───────────────────────────── init device ──────────────────────────────
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
#torch.manual_seed(0);   np.random.seed(0)

# ───────────────────────────── gen_x0(t_grid) ───────────────────────────
def gen_x00(num_samples):

    samples = np.random.normal(loc=0.1, scale=np.sqrt(0.105), size=(num_samples, 1))

    return samples
def gen_x0(t_grid, num_samples=1000):
    """
    Retourne un tensor (T, num_samples, 1) – 1-dimension par simplicité.
    """
    T = t_grid.shape[0]
    samp = torch.empty(T, num_samples, 1, device=device)
    for j, ti in enumerate(t_grid):                      
        ti_val = float(ti)                                 
        scal_t = (np.exp(2 - ti_val) + np.exp(ti_val)) / (np.exp(2) + 1)
        samp[j,:,0] = torch.randn(num_samples, device=device) * np.sqrt(0.105) * scal_t + 0.1
    return samp                                             # (T,N,1)


def wasserstein_fp(G_NN_list,              # BR passés (longueur k)
                   G_NN_star,               # BR courant   (G_k)
                   x0, m0,                    # (N,d)  échantillon initial
                   t_grid,                 # (T,1)  instants
                   p          = 1,
                   blur       = 0.05,
                   device     = "cuda"):
    """
    W_p(  m̄_k(t) , BR_k(t) )  pour t ∈ t_grid,   avec
        m̄_k = (1/(k+1)) [ δ_{x0} + Σ_{i=0}^{k-1} δ_{G_i} ].
    Retourne  np.ndarray shape (T,)  des distances.
    """

    T       = t_grid.shape[0]
    N, d    = x0.shape
    k       = len(G_NN_list)
    loss_fn = SamplesLoss("sinkhorn", p=p, blur=blur, backend="tensorized")

    # ---------- pré-calcul des trajectoires BR_k -------------------
    t_big = t_grid.repeat_interleave(N, dim=0)          # (T*N,1)
    x_rep = x0.repeat(T, 1)                             # (T*N,d)
    br_k  = G_NN_star.to(device)        # (T,N,d)

    # ---------- pré-calcul des trajectoires passées ----------------
    past_traj = []                                          # liste (k,T,N,d)
    for G in G_NN_list:
        with torch.no_grad():
            past = G(t_big, x_rep).view(T, N, d)
        past_traj.append(past)
    if past_traj:
        past_traj = torch.stack(past_traj, dim=0)           # (k,T,N,d)

    # ---------- distances -----------------------------------------
    dists = torch.empty(T, device="cpu")
    for j in range(T):
        # population : x0 + BR passés (chacun N pts)
        parts = [m0]                                        # (N,d)
        if k:
            parts.append(past_traj[:, j, :, :].reshape(k*N, d))
        X = torch.cat(parts, dim=0)                         # ((k+1)*N,d)
        Y = br_k[j]                                         # (N,d)
        dists[j] = loss_fn(X, Y).cpu()

    return dists.numpy()        # shape (T,)

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
M0 = np.load('m0.npz')
m0 =  torch.tensor(M0['x'], dtype=torch.float32, device=device)
t_grid = torch.linspace(0, 1, 21, device=device).unsqueeze(1)          # (T,1)
act_func = lambda x: torch.relu(x)



G_history = []                       # pile des BR


G_star = gen_x0(t_grid)
for k in range(20):
    
    x0_ini = gen_x00(1000)
    x0_ini = torch.tensor(x0_ini, dtype=torch.float32, device=device)
    mu  = x0_ini.mean(0);   std = x0_ini.std(0)
    G_cfg = dict(dim=1, ns=128, act_func=act_func, hh=0.5,
                 device=device, mu=mu, std=std, TT=1)

    W_k = wasserstein_fp(G_history, G_star,
                         x0_ini, m0, t_grid,
                         p=1, blur=0.5, device=device)

    print(f"q {k:2d} │ "
          f"L1={np.linalg.norm(W_k,1):.3e} │ "
          f"L2={np.linalg.norm(W_k,2):.3e} │ "
          f"L∞={W_k.max():.3e}")
    
    G_k = G_Net(**G_cfg).to(device)
    sd  = torch.load(f"G_NNN_round0_n{k}.pth", weights_only=True)
    G_k.load_state_dict(sd)

    G_history.append(copy.deepcopy(G_k))   # empiler
```
